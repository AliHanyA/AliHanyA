# ***Introduction to data science***



### **Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### ***import data***

```python
my_data = pd.read_csv('/content/diabetes.csv')
my_data.head(10)
```

## ***Analyzing tha data***


```python
# data shape
print(f"Shape of my data  y = {my_data.shape[0]} , x = {my_data.shape[1]}")

# show data info
print('='*50+"\n"+"="*50)

my_data.info()

print('='*50+"\n"+"="*50)

# checking missing & duplicate
print(f"data null = {my_data.isnull().sum().sum()}")
print(f"data duplicate = {my_data.duplicated().sum()}")

```

## visualization



```python
# To see the distribution of blood glucose levels
plt.title('Distribution of blood glucose levels', fontsize=16)
plt.xlabel('glucose level (mg/dL)', fontsize=12)
plt.ylabel('repetition', fontsize=12)

plt.grid(axis='y', alpha=0.3)


sns.histplot(data=my_data, x='Glucose')
plt.show()
```

```python
# To understand the distribution of body mass index
plt.title('distribution of body mass index', fontsize=16, pad=20)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('repetition', fontsize=12)

plt.grid(axis='y', alpha=0.3)


sns.histplot(data=my_data, x='BMI')
plt.show()
```

```python
# To compare glucose levels between diabetic and non-diabetic individuals
plt.title('Glucose SV Outcome', fontsize=16, pad=20)
plt.xlabel('Outcome', fontsize=12)
plt.ylabel('Glucose', fontsize=12)

plt.grid(axis='y', alpha=0.3)



sns.boxplot(data=my_data, x='Outcome', y='Glucose')
plt.show()
```

```python
# To observe BMI differences between both groups

plt.title('BMI SV Outcome', fontsize=16, pad=20)
plt.xlabel('Outcome', fontsize=12)
plt.ylabel('BMI', fontsize=12)

plt.grid(axis='y', alpha=0.3)



sns.boxplot(data=my_data, x='Outcome', y='BMI')
plt.show()
```

```python
# To examine the relationship between blood glucose and body mass

plt.title('relationship between blood glucose and body mass', fontsize=16, pad=20)
plt.xlabel('Glucose', fontsize=12)
plt.ylabel('BMI', fontsize=12)

plt.grid(axis='y', alpha=0.3)


sns.scatterplot(data=my_data, x='Glucose', y='BMI')
plt.show()
```

```python
sns.scatterplot(x='Glucose', y='Insulin', hue='Outcome', data=my_data, palette={0: 'blue', 1: 'orange'})
plt.title('Scatter Plot of Glucose vs Insulin by Outcome')
plt.xlabel('Glucose')
```

```python
# To explore whether age is related to insulin levels

plt.title('explore whether age is related to insulin levels', fontsize=16, pad=20)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Insulin', fontsize=12)

plt.grid(axis='y', alpha=0.3)


sns.scatterplot(data=my_data, x='Age', y='Insulin')
plt.show()
```

```python

# skin thickness distribution
sns.histplot(data=my_data, x='SkinThickness', color='red', kde=True, bins=20)
plt.title('(SkinThickness)')
plt.xlabel('(mm)')
plt.ylabel('repetition')
plt.grid(axis='y', alpha=0.3)
```

```python
# blood pressure distribution
sns.histplot(data=my_data, x='BloodPressure', kde=True, bins=20,color='black')
plt.title('(BloodPressure)')
plt.xlabel('(mmHg)')
plt.ylabel('repetition')
plt.grid(axis='y', alpha=0.3)
```

```python
# Distribution of the genetic diabetes function
sns.histplot(data=my_data, x='DiabetesPedigreeFunction', kde=True, bins=20,color='purple')
plt.title('(DiabetesPedigreeFunction)')
plt.xlabel('Genetic function value')
plt.ylabel('repetition')
plt.grid(axis='y', alpha=0.3)
```

```python
# hitmap
sns.heatmap(my_data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.pairplot(my_data)
plt.suptitle('Pairwise Relationships in Diabetes Dataset', y=1.02,fontsize=20)
plt.show()
```
